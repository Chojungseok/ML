```python
import pandas as pd

data_path = 'data/Fish.csv'
df = pd.read_csv(data_path)
```

```python
# isin(): 2개 이상의 데이터 유무 파악
cond = df['Species'].isin(['Bream', 'Smelt'])
fish_data = df.loc[cond, ['Length2', 'Weight']]
fish_data.values.tolist() # values : dataframe을 numpy array 형태로 바꾸어 진다
```

```python
fish_target = [1] * 35 + [0] * 14
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
train_input = fish_data[:35]
train_target = fish_target[:35]

test_input = fish_data[35:]
test_target = fish_target[35:]
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

## iris

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
# setosa, versicolor
# petal_length, petal_width
iris
```

```python
cond = iris['species'].isin(['setosa', 'versicolor'])
iris_data = iris.loc[cond, ['petal_length', 'petal_width']]
iris_data.values.tolist()
```

```python
iris_target = [0] * sum(iris['species'] == 'setosa') + [1] * sum(iris['species'] == 'versicolor')
iris_target
```

```python
np.random.seed(42)
index = np.arange(100)
np.random.shuffle(index)
print(index)
```

```python
input_arr = np.array(iris_data)
target_arr = np.array(iris_target)
```

```python
train_input = input_arr[index[:70]]
train_target = target_arr[index[:70]]

test_input = input_arr[index[70:]]
test_target = target_arr[index[70:]]
```

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
plt.scatter(train_input[:,0],train_input[:,1])
plt.scatter(test_input[:,0],test_input[:,1])
```
