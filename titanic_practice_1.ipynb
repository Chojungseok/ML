```python
import pandas as pd
```

```python
file_path = 'data/'
df = pd.read_csv(file_path+'train.csv')
```

```python
df
```

```python
target_df = df.loc[:,['Survived']]
train_df = df.loc[:,['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']]
```

```python
train_df
```

```python
train_df.info()
```

```python
train_df[train_df['Embarked'].isna()]
```

```python
train_df['Embarked'].value_counts()
```

```python
train_df['Embarked'].fillna('S', inplace=True)
```

```python
train_df['Embarked'].value_counts()
```

---

```python
train_df[train_df['Age'].isna()]
```

```python
train_df['Age'].mean().round()
```

```python
train_df['Age'].fillna(30.0, inplace=True)
```

```python
train_df.drop('Cabin', axis=1, inplace=True)
train_df
```

```python
train_df = pd.get_dummies(train_df, columns=['Sex'])
```

```python
train_df['Sex_female'] = train_df['Sex_female'].astype(int)
train_df['Sex_male'] = train_df['Sex_male'].astype(int)
```

```python
train_df
```

```python
import numpy as np
len(train_df['Ticket'].unique())
```

```python
train_df.drop('Ticket', axis=1, inplace=True)
train_df
```

```python
from sklearn.preprocessing import LabelEncoder

lable = LabelEncoder()
```

```python
train_df['Embarked'] = lable.fit_transform(train_df['Embarked'])
```

```python
train_df['Embarked'].value_counts()
```

```python
train_df.info()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


tr_df, test_df, tr_target, test_target = train_test_split(train_df, target_df, test_size=0.2, random_state=42)

lg = LogisticRegression()
```

```python
lg.fit(tr_df, tr_target)
```

```python
print(lg.score(tr_df, tr_target))
print(lg.score(test_df, test_target))
```

```python
lg.predict_proba(tr_df[:10])
```

```python
lg.predict(tr_df[:10])
```

```python
tr_target[:10]
```

---

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(random_state=42)
```

```python
rfc.fit(tr_df, tr_target)
```

```python
print(rfc.score(tr_df, tr_target))
print(rfc.score(test_df, test_target))
```

```python
### oob사용
rfc = RandomForestClassifier(oob_score=True)

rfc.fit(train_df, target_df)
rfc.oob_score_
```

```python
from sklearn.model_selection import GridSearchCV

parms = {
    'n_estimators': np.arange(50, 250, 10),
    'max_depth' : np.arange(1,25,5)
}

gs = GridSearchCV(rfc, parms, n_jobs=-1)
```

```python
gs.fit(train_df,target_df)
```

```python
print(rfc.score(train_df,target_df))
```

```python
rfc.oob_score_
```
