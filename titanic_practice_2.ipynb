```python
import pandas as pd
```

```python
# pd.set_option('display.max_rows')
```

```python
file_path = 'data/'
df = pd.read_csv(file_path+'train.csv')
```

```python
df
```

```python
target_df = df.loc[:,['Survived']]
train_df = df.loc[:,['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']]
```

```python
train_df
```

```python
train_df.info()
```

```python
train_df
```

```python
train_df.drop(['PassengerId','Ticket', 'Embarked'], axis=1, inplace=True)
```

```python
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].mean())
```

```python
train_df['Age'] = train_df['Age'].astype(int)
```

```python
train_df.info()
```

```python
# train_df['Cabin'] = train_df['Cabin'].fillna('N', inplace=True)
```

```python
train_df
```

```python
train_df['SibSp'].value_counts()
```

```python
sibsp = []
for index, row in train_df[['SibSp']].iterrows():
    if row[0] != 0:
        sibsp.append(1)
    else:
        sibsp.append(0)
    
    
```

```python
train_df['SibSp'] = sibsp
```

```python
train_df['SibSp'].value_counts()
```

```python
# pd.set_option('display.max_rows')
# train_df
pd.reset_option('display.max_rows')
```

```python
train_df.drop('Cabin', axis = 1, inplace=True)
```

```python
train_df
```

```python
train_df = pd.get_dummies(train_df, columns=['Sex'])
```

```python
train_df['Sex_female'] =train_df['Sex_female'].astype(int)
train_df['Sex_male'] =train_df['Sex_male'].astype(int) 
```

```python
train_df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
lg  = LogisticRegression()
```

```python
tr_df, te_df, tr_target, te_target = train_test_split(train_df, target_df, random_state=42)
```

```python
lg.fit(tr_df,tr_target)
```

```python
print(lg.score(tr_df,tr_target))
print(lg.score(te_df,te_target))
```
