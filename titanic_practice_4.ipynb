```python
import pandas as pd
```

```python
file_path = 'data/'
df = pd.read_csv(file_path+'train.csv')
```

```python
df
```

```python
df.drop(['PassengerId', 'Name', 'Ticket'], axis = 1, inplace = True)
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
df.info()
```

```python
df['Cabin'] = df['Cabin'].fillna('N')
```

```python
# df.drop('Cabin', axis = 1, inplace = True)
```

```python
df
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df
```

```python
floor = []
for index, row in df[['Cabin']].iterrows():
    floor.append(row[0][0])
```

```python
df['floor'] = floor
```

```python
df.drop('Cabin', axis = 1, inplace = True)
```

```python
df
```

```python
target_df = df[['Survived']]
train_df = df.loc[:,['Pclass', 'Sex','Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'floor']]
```

```python
train_df
```

```python
train_df = pd.get_dummies(train_df, columns=['Sex', 'Embarked', 'floor'], dtype=int)
```

```python
train_df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()


clf = RandomForestClassifier(random_state=42)
trian_input, test_input, train_target, test_target = train_test_split(train_df, target_df, test_size=0.2,random_state=42)
```

```python
trian_input = scaler.fit_transform(trian_input)
test_input = scaler.transform(test_input)
```

```python
clf.fit(trian_input, train_target)
```

```python
print(clf.score(trian_input, train_target))
print(clf.score(test_input, test_target))
```
