```python
import pandas as pd
```

```python
file_path = '/Users/chojungseok/Desktop/code/DAMF2/ML/data/Fish.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
df['Species'].unique()
```

```python
fish_input = df[['Weight', 'Length2', 'Length3', 'Height', 'Width']]
fish_input.head()
```

```python
fish_target = df[['Species']]
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_scaled, train_target)
```

```python
print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

```python
kn.predict(test_scaled)
```

```python
# predict_proba: 분류한 확률을 볼 수 있다
kn.predict_proba(test_scaled)
```

# 로지스틱 회귀
$z = a * Weight + b * Length + c * Diagonal + d * Height + e * Width + f$  
회귀라고 적혀있지만 분류 모델임  
일단 $z$값을 구해서 어느 분류에 속할지 확률을 구하는 것임  

단 $z$는 엄청 큰/작은 숫자가 나올 것임 따라서 확률로 변환해주는 과정이 필요

*시그모이드 함수*
- 양의 무한대로 갈때 1에 수렴한다
- 음의 무한대로 갈때 0에 수렴한다  
**0~1사이의 확률을 갖는다**

```python
train_scaled
```

```python
bream_smelt_indexes = (train_target == 'Bream') | (train_target == 'Smelt')

train_bs = train_scaled[bream_smelt_indexes['Species']]
target_bs = train_target[bream_smelt_indexes['Species']]
# target_bs
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_bs, target_bs)
```

```python
lr.predict(train_bs[:5])
```

```python
lr.predict_proba(train_bs[:5])
```

```python
lr.coef_
```

```python
lr.decision_function(train_bs[:5]) # 값들을 시그모이드 함수에 넣으면 이제 확율로 나오게 된다
```

```python
lr = LogisticRegression(C = 20, max_iter=1000)

lr.fit(train_scaled, train_target)
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

로지스틱 회귀분석의 hyperparameter  
max_iter, C  



```python
lr.predict_proba(test_scaled[:5]).round(3)
# 7종류의 생선들 중에 1개에 속할 확률들
```

```python
lr.coef_.shape
```

다중 회귀에서 로지스틱 함수를 사용하게 된다면 분류의 클래스 개수만큼 회귀식이 생성이 되고 그 회귀식을 모두 합하면 1을 넘어가버리므로 다중회귀에서는 로지스틱을 쓰지 않는다.  
-> softmax함수 사용  
로지스틱 함수를 한 다음에 전체를 softmax함수를 사용해주면 전체 합이 1안으로 들어오는 확률로 변환해준다
